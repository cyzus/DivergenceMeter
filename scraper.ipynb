```python
import requests
import urllib.request
import time
from bs4 import BeautifulSoup
```

```python
url = 'http://www.piano-midi.de/midi_files.htm'
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")
links = dict()
for one_a_tag in soup.findAll('a'):  #'a' tags are for links
    try:
        link = one_a_tag['href']
        # if link in links:
        #     continue
        links[link] = 1
        download_url = 'http://www.piano-midi.de'+ '/' + link
        
        url = download_url
        #print(download_url)
        response = requests.get(url)
        subsoup = BeautifulSoup(response.text, "html.parser")
        for sub in subsoup.findAll('a'):
            try:
                link = sub['href']
                if link in links:
                    continue
                links[link] = 1
                #print(link)
                if '.mid' in link:
                    print(link)
                    if "format" in link:
                        continue
                    download_url = 'http://www.piano-midi.de'+ '/' + link
                    print(download_url)
                    
                    sid = link.rfind('/')
                    filename = 'midi/' + link[sid+1:]
                    print(filename)
                    urllib.request.urlretrieve(download_url, filename) 
                    time.sleep(1) #pause the code for a sec
            except:
                print('terrible things happend')
    except:
        pass
    
```
